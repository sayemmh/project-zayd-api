```python
import pandas as pd
import ujson
import json
import pickle as pckl
json_surah_words = 'json-surah-words/'
```

```python
surah_df = pd.read_json('df.json')
nan_rows = surah_df[surah_df['waznType'].isnull()]
surah_df = surah_df[surah_df['waznType'].notnull()]
nan_rows.drop_duplicates(['question']).head(35)
# n_rows = nan_rows.drop_duplicates(['answer', 'question'])
# n_rows.drop_duplicates(['question'])
# surah_df.head()
```

```python
particle_df = surah_df.copy(deep=True)
mask = surah_df['waznType'].str.contains('particle')
particle_df = surah_df[mask]
for ind in particle_df.index:
    old_answer = particle_df['answer'][ind]
    getVals = list([val.lower() for val in old_answer 
               if val.isalpha() or val.isnumeric() or val.isspace()])
    result = "".join(getVals)
    particle_df['answer'][ind] = result
particle_df = particle_df.drop_duplicates(['answer','question'])
particle_df
```

```python
nonroot_particle = pd.read_json('particles.json')
for ind in nonroot_particle.index:
    old_answer = nonroot_particle['answer'][ind]
    getVals = list([val.lower() for val in old_answer 
               if val.isalpha() or val.isnumeric() or val.isspace()])
    result = "".join(getVals)
    nonroot_particle['answer'][ind] = result
nonroot_particle = nonroot_particle.drop(columns=['level_0'])
nonroot_particle = nonroot_particle.drop_duplicates(['answer','question'])
nonroot_particle
```

```python
final_particle = pd.concat([nonroot_particle, particle_df],ignore_index=True)
final_particle = final_particle[final_particle['frequency'] != 1464]
final_particle
```

```python
df2 = pd.read_json('df2.json')
df2 = df2.reset_index()
df2
```

```python
df2_grouped = df2.drop_duplicates(['question','morphCode']).reset_index()
df2_grouped = df2_grouped.groupby('answer').first().reset_index()

df2_grouped = df2_grouped.sort_values(by=['question'])

df2_grouped['Category'] = df2_grouped['morphCode']
df2_grouped['Category'] = df2_grouped['Category'].replace(to_replace='PN',value='Proper Noun')


df2_grouped['Category'] = df2_grouped['Category'].replace(to_replace='P',value='Preposition')
# df2_grouped[df2_grouped['Category']== 'Preposition']
df2_grouped['Category'] = df2_grouped['Category'].replace(to_replace='PRON',value='Pronoun')

proper_nouns = pd.DataFrame()
preps = pd.DataFrame()
pronouns = pd.DataFrame()
for i in range(len(df2)):
    code_arr = df2.loc[i,'morphCode'].split(",")
    if 'PN' in code_arr:
        proper_nouns = proper_nouns.append(df2.loc[i])
    elif 'P' in code_arr:
        preps = preps.append(df2.loc[i])
    elif 'PRON' in code_arr:
        pronouns = pronouns.append(df2.loc[i])
proper_nouns
```
