```python
import pandas as pd
import json
import csv
import ujson
```

```python
df_list = list()
for i in range(1,115):
    temp_df = pd.read_json('ayahRoots/' + str(i) + '.json')
    df_list += [temp_df]
ayah_df = pd.concat(df_list)
ayah_df = ayah_df.loc[ayah_df['Difficulty'] != 0] # gets rid of ayahs without any root words (alif lam meem)
ayah_df = ayah_df.sort_values(by=['Difficulty'], ascending = False) # easiest at top
ayah_df
```

```python
with open('word-scores.json') as f:
    data = json.load(f)
word_df = pd.DataFrame.from_dict(data, orient='index', columns=['Score'])
word_df = word_df.sort_values(by=['Score'], ascending = False) # easiest at top
word_df
```

```python
# for each word in top x words, count how many of the top y ayahs they appear in
def find_ayah(word_df, ayah_df, end_word, end_ayah):
    word_count_list = list()
    for ind in word_df[:end_word].index:
        word_count = 0
        for j in range(end_ayah):
            id_list = ayah_df.iloc[j]['word_ids']
            word_count += id_list.count(ind)
        word_count_list.append(word_count)
#         print(str(ind) + " count: " + str(word_count))
    return word_count_list
```

```python
order = list()
while word_df.size > 0:
    end_word = word_df.size
    end_ayah = ayah_df.size
    if end_word > 50:
        end_word = 50
    if end_ayah > 500:
        end_ayah = 500
        
    word_count_list = find_ayah(word_df, ayah_df, end_word, end_ayah)
    max_index = word_count_list.index(max(word_count_list))
    index_list = word_df.index.values
    drop_index = index_list[max_index]
    order.append(drop_index)
    print(drop_index)
    word_df = word_df.drop(drop_index)
    for ind in ayah_df.index:
        id_list = ayah_df.iloc[ind]['word_ids']
        id_list[:] = [x for x in id_list if x != drop_index]
        if len(id_list) == 0:
            ayah_df.drop(ind)
        else:
            ayah_df.iloc[ind]['word_ids'] = id_list
```
