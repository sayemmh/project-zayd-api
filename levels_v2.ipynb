```python
# for words that don't have arabic ayahs for context, this finds the next closest word
def getArabicAyah(surahnum, ayahnum, wordnum, df):
#     surahnum = 2
#     ayahnum = 286
#     wordnum = 1
    ayahSearch = True
    count = 1
    print(str(surahnum)+":"+str(ayahnum)+":"+str(wordnum))
    while ayahSearch:
        if count >= 5:
            ayahSearch = False 
        if ((df['surahnum'] == surahnum) & (df['ayahnum'] == ayahnum) & (df['wordnum'] == wordnum-count)).any():

            arabicAyah = df[((df['surahnum'] == surahnum) & (df['ayahnum'] == ayahnum) & (df['wordnum'] == wordnum-count))].iloc[0]['arabicAyah']
            return arabicAyah
            ayahSearch = False
        elif ((df['surahnum'] == surahnum) & (df['ayahnum'] == ayahnum) & (df['wordnum'] == wordnum+count)).any():
            arabicAyah = df[((df['surahnum'] == surahnum) & (df['ayahnum'] == ayahnum) & (df['wordnum'] == wordnum+count))].iloc[0]['arabicAyah']
            return arabicAyah
            ayahSearch = False
        count += 1 
```
